```python
correct = []
incorrect = []
for x in full_df['combo'].unique().tolist():
    if type(x) != float:
        a, v = x.split('_')
        if a == v:
            correct.append(v)
            print(f'Speaker {v[1:]} is correct')
        else:
            incorrect.append((v,a))
            print(f"Speaker {v[1:]}'s videos need to be pointed to speaker {a[1:]}'s alignments")
```
